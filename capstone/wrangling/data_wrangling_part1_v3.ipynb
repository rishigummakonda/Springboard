```python
state_codes=[]
state=""
county_codes=[]
county=""
tract_codes=[]
tract=""

for code in nyc_codes:
    code=str(code)
    state=code[:2]
    county=code[2:5]
    tract=code[5:11]
    state_codes.append(state)
    county_codes.append(county)
    tract_codes.append(tract)
```
